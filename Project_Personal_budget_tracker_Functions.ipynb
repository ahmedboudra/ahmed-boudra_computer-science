```python
# Personal Budget Tracker

# Data storage
income_list = []
expense_list = []

# Function to add income
def add_income():
    description = input("Enter income description (e.g., Salary, Freelance): ")
    try:
        amount = float(input("Enter income amount: "))
        income_list.append({"description": description, "amount": amount})
        print("Income added successfully!\n")
    except ValueError:
        print("Invalid amount. Please enter a number.\n")

# Function to add expense
def add_expense():
    description = input("Enter expense description (e.g., Rent, Groceries): ")
    try:
        amount = float(input("Enter expense amount: "))
        category = input("Enter expense category (e.g., Food, Utilities): ")
        expense_list.append({"description": description, "amount": amount, "category": category})
        print("Expense added successfully!\n")
    except ValueError:
        print("Invalid amount. Please enter a number.\n")

# Function to view budget summary
def view_summary():
    total_income = sum(item["amount"] for item in income_list)
    total_expense = sum(item["amount"] for item in expense_list)
    balance = total_income - total_expense

    print("\n===== Budget Summary =====")
    print(f"Total Income: ${total_income:.2f}")
    print(f"Total Expenses: ${total_expense:.2f}")
    print(f"Balance: ${balance:.2f}\n")

# Function to delete an entry
def delete_entry():
    choice = input("Delete 'income' or 'expense' entry? ").lower()
    description = input("Enter the description of the entry to delete: ")

    if choice == "income":
        for entry in income_list:
            if entry["description"].lower() == description.lower():
                income_list.remove(entry)
                print("Income entry deleted.\n")
                return
        print("Income entry not found.\n")
    elif choice == "expense":
        for entry in expense_list:
            if entry["description"].lower() == description.lower():
                expense_list.remove(entry)
                print("Expense entry deleted.\n")
                return
        print("Expense entry not found.\n")
    else:
        print("Invalid choice. Please enter 'income' or 'expense'.\n")

# Function to modify an entry
def modify_entry():
    choice = input("Modify 'income' or 'expense' entry? ").lower()
    description = input("Enter the description of the entry to modify: ")

    if choice == "income":
        for entry in income_list:
            if entry["description"].lower() == description.lower():
                entry["description"] = input("Enter new description: ")
                entry["amount"] = float(input("Enter new amount: "))
                print("Income entry modified.\n")
                return
        print("Income entry not found.\n")
    elif choice == "expense":
        for entry in expense_list:
            if entry["description"].lower() == description.lower():
                entry["description"] = input("Enter new description: ")
                entry["amount"] = float(input("Enter new amount: "))
                entry["category"] = input("Enter new category: ")
                print("Expense entry modified.\n")
                return
        print("Expense entry not found.\n")
    else:
        print("Invalid choice. Please enter 'income' or 'expense'.\n")

# Function to list all entries
def list_entries():
    print("\n===== Income Entries =====")
    if income_list:
        for entry in income_list:
            print(f"- {entry['description']}: ${entry['amount']:.2f}")
    else:
        print("No income entries found.")

    print("\n===== Expense Entries =====")
    if expense_list:
        for entry in expense_list:
            print(f"- {entry['description']} (${entry['amount']:.2f}) - Category: {entry['category']}")
    else:
        print("No expense entries found.")
    print()

# Main function with interactive menu
def main():
    while True:
        print("===== Personal Budget Tracker Menu =====")
        print("1. Add Income")
        print("2. Add Expense")
        print("3. View Budget Summary")
        print("4. Delete an Entry")
        print("5. Modify an Entry")
        print("6. List All Entries")
        print("7. Exit")

        choice = input("Select an option (1-7): ")

        if choice == "1":
            add_income()
        elif choice == "2":
            add_expense()
        elif choice == "3":
            view_summary()
        elif choice == "4":
            delete_entry()
        elif choice == "5":
            modify_entry()
        elif choice == "6":
            list_entries()
        elif choice == "7":
            print("Exiting Personal Budget Tracker. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 7.\n")

# Run the program
main()
```
