```python
# Task 1: Assigning Values to Variables
integer_var = 10            # Integer
float_var = 3.14            # Floating-point number
string_var = "Python"       # String
print(integer_var, float_var, string_var)

# Task 2: Data Types
age = 25
print("Data type of age:", type(age))  # <class 'int'>
age = 25.0
print("Data type of age (as float):", type(age))  # <class 'float'>

is_student = True
print("Data type of is_student:", type(is_student))  # <class 'bool'>

# Task 3: Basic Input and Output
name = input("What is your name? ")
print("Hello,", name + "!")

user_age = int(input("Enter your age: "))
if user_age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")

# Task 4: Converting Data Types
number_str = input("Enter a number: ")
number_int = int(number_str)
print("Squared value:", number_int ** 2)

float_str = "4.56"
float_num = float(float_str)
print("Double the value:", float_num * 2)

# Task 5: Arithmetic Operators
print("45 + 67 =", 45 + 67)
print("156 - 89 =", 156 - 89)
print("12 * 7 =", 12 * 7)
print("100 / 4 =", 100 / 4)
print("10 % 3 =", 10 % 3)
print("3 ** 4 =", 3 ** 4)

# Task 6: Comparison Operators
a = 10
b = 20
print("a == b:", a == b)
print("a != b:", a != b)
print("a < b:", a < b)
print("a > b:", a > b)
print("a <= b:", a <= b)
print("a >= b:", a >= b)

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
if num1 >= num2:
    print("First number is greater than or equal to the second.")
else:
    print("First number is less than the second.")

# Task 7: Logical Operators
num = int(input("Enter a number: "))
if num > 10 and num < 20:
    print("Number is between 10 and 20.")
else:
    print("Number is NOT between 10 and 20.")

num = int(input("Enter another number: "))
if num < 5 or num > 50:
    print("Number is either less than 5 or greater than 50.")
else:
    print("Number is between 5 and 50.")

# Task 8: Conditional Statements
age = int(input("Enter your age: "))
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")

number = int(input("Enter a number: "))
if number % 3 == 0 and number % 5 == 0:
    print("FizzBuzz")
elif number % 3 == 0:
    print("Fizz")
elif number % 5 == 0:
    print("Buzz")
else:
    print(number)

# Task 9: Loops
print("For loop from 1 to 10:")
for i in range(1, 11):
    print(i)

print("Even numbers from 2 to 20 using while loop:")
i = 2
while i <= 20:
    print(i)
    i += 2

n = int(input("Enter a number n: "))
print("Numbers divisible by 4 from 1 to", n)
for i in range(1, n + 1):
    if i % 4 == 0:
        print(i)
```
