```python
# Movie Library System â€“ Friendly Version

# Step 1: Initialize the movie library
movie_library = []

# Step 2: Add some sample movies
movie_library.append({
    "Title": "Inception",
    "Year": 2010,
    "Genres": ["Action", "Sci-Fi"],
    "Rating": (4.5, 1500),
    "Reviews": {
        "Alice": "Mind-blowing concept and visuals!",
        "Bob": "Nolan's best work."
    }
})

movie_library.append({
    "Title": "Titanic",
    "Year": 1997,
    "Genres": ["Romance", "Drama"],
    "Rating": (4.7, 2000),
    "Reviews": {
        "Clara": "Emotional and beautifully made.",
        "Daniel": "A timeless classic."
    }
})

movie_library.append({
    "Title": "The Matrix",
    "Year": 1999,
    "Genres": ["Action", "Sci-Fi"],
    "Rating": (4.8, 1800),
    "Reviews": {
        "Eve": "A revolutionary film.",
        "Frank": "Changed the genre forever."
    }
})


# View all movies
def view_all_movies():
    print("\nğŸ¬ All Available Movies:")
    for movie in movie_library:
        print(f"\nğŸ“Œ Title: {movie['Title']}")
        print(f"ğŸ“… Released: {movie['Year']}")
        print(f"ğŸ­ Genres: {', '.join(movie['Genres'])}")
        print(f"â­ Rating: {movie['Rating'][0]} (based on {movie['Rating'][1]} reviews)")

# Search movies by genre
def search_by_genre(genre):
    print(f"\nğŸ” Looking for movies in the '{genre}' genre...")
    found = False
    for movie in movie_library:
        if genre.capitalize() in [g.capitalize() for g in movie['Genres']]:
            print(f"- {movie['Title']}")
            found = True
    if not found:
        print("ğŸ˜• No movies found in that genre.")

# View details of a movie
def view_movie_details(title):
    for movie in movie_library:
        if movie['Title'].lower() == title.lower():
            print(f"\nğŸ“½ï¸ Full Details for '{movie['Title']}':")
            print(f"ğŸ“… Year: {movie['Year']}")
            print(f"ğŸ­ Genres: {', '.join(movie['Genres'])}")
            print(f"â­ Rating: {movie['Rating'][0]} (from {movie['Rating'][1]} reviews)")
            print("ğŸ“ User Reviews:")
            for reviewer, review in movie['Reviews'].items():
                print(f"  - {reviewer}: \"{review}\"")
            return
    print("ğŸš« Sorry, we couldnâ€™t find that movie.")

# Add a new review to a movie
def add_review(title, reviewer, review_text):
    for movie in movie_library:
        if movie['Title'].lower() == title.lower():
            movie['Reviews'][reviewer] = review_text
            current_total = movie['Rating'][0] * movie['Rating'][1]
            new_total_reviews = movie['Rating'][1] + 1
            new_avg_rating = (current_total + 5.0) / new_total_reviews
            movie['Rating'] = (round(new_avg_rating, 2), new_total_reviews)
            print(f"âœ… Thank you {reviewer}, your review has been added to '{title}'!")
            return
    print("ğŸš« Movie not found. Try checking the title again.")

# Calculate average rating across all movies
def calculate_average_rating():
    total_score = 0
    total_reviews = 0
    for movie in movie_library:
        total_score += movie['Rating'][0] * movie['Rating'][1]
        total_reviews += movie['Rating'][1]
    if total_reviews > 0:
        avg = total_score / total_reviews
        print(f"\nğŸ“Š Average rating across all movies: {round(avg, 2)} stars")
    else:
        print("ğŸ¤· No reviews found to calculate average.")

# Main menu for user interaction
def main_menu():
    print("ğŸ‰ Welcome to the Movie Library System!")
    while True:
        print("\n----------------------------")
        print("Please choose an option:")
        print("1. View all movies")
        print("2. Search movies by genre")
        print("3. View detailed info about a movie")
        print("4. Add your own review")
        print("5. See the average rating of all movies")
        print("6. Exit the library")
        print("----------------------------")

        choice = input("Enter your choice (1â€“6): ").strip()

        if choice == '1':
            view_all_movies()
        elif choice == '2':
            genre = input("ğŸ­ Enter a genre to search for: ")
            search_by_genre(genre)
        elif choice == '3':
            title = input("ğŸ¬ Enter the movie title: ")
            view_movie_details(title)
        elif choice == '4':
            title = input("ğŸ¬ Which movie do you want to review? ")
            reviewer = input("ğŸ§‘â€ğŸ’» What's your name? ")
            review_text = input("ğŸ“ Write your review: ")
            add_review(title, reviewer, review_text)
        elif choice == '5':
            calculate_average_rating()
        elif choice == '6':
            print("ğŸ‘‹ Thanks for using the Movie Library System. Goodbye!")
            break
        else:
            print("âš ï¸ Invalid choice. Please enter a number from 1 to 6.")

# Start the program
main_menu()
```
